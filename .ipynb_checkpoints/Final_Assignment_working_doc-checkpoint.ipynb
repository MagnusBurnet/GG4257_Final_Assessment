```python
merged_data = shape.merge(crash_data, left_on='LSOA11CD', right_on='LSOA_of_Accident_Location', how='left')
```

```python
merged_data.head()
```

```python
# Assuming merged_data is your DataFrame
# file_path = '/path/to/your/data.csv' # For example purposes
# merged_data = pd.read_csv(file_path)

# List of variables to transform
variables = ['Weather_Conditions', 'Accident_Severity', 'Number_of_Vehicles', 
             'Number_of_Casualties', 'Road_Surface_Conditions', 
             'Light_Conditions', 'Did_Police_Officer_Attend_Scene_of_Accident', 
             'Urban_or_Rural_Area']

# Function to group by LSOA and another variable, then pivot
def pivot_counts(data, group_var):
    counts = data.groupby(['LSOA_of_Accident_Location', group_var]).size().reset_index(name='Count')
    pivot = counts.pivot(index='LSOA_of_Accident_Location', columns=group_var, values='Count').fillna(0)
    # Ensure the new column names are unique across different variables
    pivot.columns = [f"{group_var}_{col}" for col in pivot.columns]
    return pivot.reset_index()

# Apply the function for each variable and merge the results
final_df = None
for var in variables:
    pivot_df = pivot_counts(merged_data, var)
    if final_df is None:
        final_df = pivot_df
    else:
        final_df = final_df.merge(pivot_df, on='LSOA_of_Accident_Location', how='outer')

# Fill NaNs that result from outer join with 0s
final_df = final_df.fillna(0)

# Optionally, convert all counts to integers
for col in final_df.columns:
    if col != 'LSOA_of_Accident_Location':
        final_df[col] = final_df[col].astype(int)

# Display the first few rows of the final DataFrame
final_df.head()
```

```python
numeric_columns = final_df.select_dtypes(include='int')
z_score_df = (numeric_columns - numeric_columns.mean()) / numeric_columns.std(ddof=0)
z_score_df.head()
```

```python
corr = z_score_df.corr()
corr.style.background_gradient(cmap='coolwarm')
```

```python
#Plot histograms of some of the key attributes
import seaborn as sns
import matplotlib.pyplot as plt
#For ease of interpretation,
#the following code plots the same table however highlights only the variables with >0.75 correlation.
threshold = 0.8 

highly_correlated = (corr.abs() > threshold) & (corr.abs() < 1.0)

#Use heat map to help differentiate between highly correlated relationships
plt.figure(figsize=(10, 8))
sns.heatmap(highly_correlated, cmap='coolwarm', cbar=False, annot=True)

plt.title('Highly Correlated Variables')
plt.show()
```
